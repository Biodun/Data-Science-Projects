###Interactive visualizations

Let's use IPython Notebook's interactive widgets to help us explore the distribution of trip durations. __You will have to download this notebook and have IPython Notebook v2.x installed on your computer before you can interact with this widget__.

````python
def plot_duration_distribution(min_trip_length, max_trip_length):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 3))
    #Filter the trip by min and max durations
    trip = data[(data.Duration_minutes >= min_trip_length) & (data.Duration_minutes < max_trip_length)]
    n_trips = len(trip)
    #Plot a histogram of the data
    trip.Duration_minutes.plot(kind='hist', bins=100, ax=ax[0], 
                       title='Trip duration Histogram\nTime duration: {} to {} minutes\nTotal of {} trips'.format(
                        min_trip_length, max_trip_length, n_trips))
    #Plot a boxplot of the data
    trip.Duration_minutes.plot(kind='box', ax=ax[1], title='Boxplot of trip duration')
````

````python
interact(plot_duration_distribution, min_trip_length=widgets.IntSliderWidget(min=1, max=299, step=1),
         max_trip_length=widgets.IntSliderWidget(min=1, max=300, step=1, value=100))
````

### What percentage of trips start and end at the same station?

````python
trips_which_start_and_end_in_diff_locs = data[data['Start Terminal'] != data['End Terminal']]
trips_startin_and_ending_in_same_loc = data[data['Start Terminal'] == data['End Terminal']]
n_trips_to_different_station = len(trips_startin_and_ending_in_same_loc)
print('Fraction of trips starting and ending at the same station: {:.03}%'.format(
                                                    (n_trips_to_different_station/float(n_trips))*100))
````

### Let's compare the distribution of trips originating and terminating at the same station with those that start and terminate at different stations 

````python
plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 5))
trips_which_start_and_end_in_diff_locs['Duration_minutes'].plot(kind='hist', ax=axes[0][0], 
                                                                label='Trips starting and ending in different locations',
                                                                bins=200, legend=True, fontsize=MEDIUM_FONTSIZE)
trips_which_start_and_end_in_diff_locs.boxplot('Duration_minutes', ax=axes[0][1], fontsize=MEDIUM_FONTSIZE)
trips_startin_and_ending_in_same_loc['Duration_minutes'].plot(kind='hist', ax=axes[1][0], 
                                                                label='Trips starting and ending in same location',
                                                                bins=200, legend=True, color='g'
                                                                , fontsize=MEDIUM_FONTSIZE)
trips_startin_and_ending_in_same_loc.boxplot('Duration_minutes', ax=axes[1][1], fontsize=LARGE_FONTSIZE)
for i in (0,1):
    #Show only trips less than 400 mins long
    axes[i][0].set_xlabel('Trip duration (minutes)', fontsize=MEDIUM_FONTSIZE)
    axes[i][0].set_ylabel('# of trips', fontsize=MEDIUM_FONTSIZE)
    axes[i][0].set_xlim(0, 400)
    axes[i][1].set_ylim(0, 300)
axes[0][1].set_ylim(0, 80)
````

* The charts above tell us that the vast majority of trips are made between different stations, and they typically last for around 13 minutes. On the other hand, trips starting ending at the same station tend to be much longer with an average duration of ~76 minutes. 

* Could it be that trips that start and end at the same station are taken by people out for a leisurely ride rather than by people commuting to work? We can check the distribution of customers types based on whether the trip started and ended in the same location and the time the trip was taken. Maybe that will give us more information.

````python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(17, 4))
trips_startin_and_ending_in_same_loc['Subscription Type'].value_counts().plot(kind='barh', fontsize=MEDIUM_FONTSIZE,
                                                                              ax=axes[0])
axes[0].set_xlabel('# of riders', fontsize=MEDIUM_FONTSIZE)
axes[0].set_title('# of trips starting and ending\nat the same station', fontsize=LARGE_FONTSIZE)
trips_which_start_and_end_in_diff_locs['Subscription Type'].value_counts().plot(kind='barh', fontsize=MEDIUM_FONTSIZE,
                                                                              ax=axes[1])
axes[1].set_xlabel('# of riders', fontsize=MEDIUM_FONTSIZE)
axes[1].set_title('# of trips starting and ending\nat the different stations', fontsize=LARGE_FONTSIZE)
plt.tight_layout()
````

Customers outnumber subscribers two to one for trips that start and end at the same location. Maybe sorting trips by time will provide us with more insight into rider patterns.

### What is the most popular day by trip frequency?

We use pandas groupby and timeseries functionality to group the data by dates, days of the week and hours of the day 

````python
daily_trip_counts = sorted_data.groupby(pd.TimeGrouper('d', how='sum'))['Start Date'].apply(lambda x: x.count())
plt.figure(figsize=(17, 3))
daily_trip_counts.plot(fontsize=MEDIUM_FONTSIZE)
plt.title('Daily trip counts', fontsize=LARGE_FONTSIZE)
plt.ylabel('# of trips', fontsize=MEDIUM_FONTSIZE)
print('Most popular day: {} , {} trips'.format(daily_trip_counts.argmax(), daily_trip_counts.max()))
````

### Find bike ridership by day of the week

Each pandas timestamp object contains day of the week and hour information we can extract 

````python
#Grab a timestamp from our Dataframe
ts = sorted_data.iloc[0]['Start Date']
````

````python
print('Full timestamp: {}'.format(ts))
````

````python
#Print the day of the week
print('Day of the week: {}'.format(ts.dayofweek))
````

You can read more about Pandas timestamps on the pandas timeseries documentation __[webpage](http://pandas.pydata.org/pandas-docs/stable/timeseries.html#time-date-components)__

***

We append two new columns to our DataFrame to provide easy access to the day of the week and the hour at which the trip started

````python
day_of_week_mapping = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 
                       5: 'Saturday', 6: 'Sunday'}
#Extract the day of the week from each timestamp and append a new column containing that information
sorted_data['Day_of_the_week'] = sorted_data['Start Date'].apply(lambda x: x.dayofweek).values
#Get the hour of each trip
sorted_data['start_hour'] = sorted_data['Start Date'].apply(lambda x : x.hour).values
````

````python
sorted_data.head(2)
````

### What time are most bikes rented?

````python
plt.figure(figsize=(17, 3))
sorted_data.groupby('start_hour')['Trip ID'].count().plot(fontsize=MEDIUM_FONTSIZE, x_compat=True)
plt.title('Bike rental by hour of the day', fontsize=LARGE_FONTSIZE)
plt.ylabel('# of Bike rentals', fontsize=MEDIUM_FONTSIZE)
plt.xlabel('Starting hour (h)', fontsize=MEDIUM_FONTSIZE)
````

Most rentals occur between 8-9am and between 4-6pm daily

### What's the average duration of a trip on each day of the week?

````python
plt.figure(figsize=(17, 3))
ax = sorted_data.groupby('Day_of_the_week')['Duration_minutes'].mean().plot(fontsize=MEDIUM_FONTSIZE)
ax.set_title('Average trip duration (min)', fontsize=LARGE_FONTSIZE)
ax.set_xticklabels(day_of_week_mapping.values());
ax.set_ylabel('Trip duration\n(Minutes)', fontsize=MEDIUM_FONTSIZE)
````

Averages can hide interesting details. If we break down the average trip duration by customer type we get find that subscribers (those with annual memberships) have very different usage patterns from customers (people who buy 1-day or 3 day passes)

````python
for s in ('Subscriber', 'Customer'):
    ax = sorted_data[sorted_data['Subscription Type']==s].groupby('Day_of_the_week')['Duration_minutes'].mean().plot(
            fontsize=MEDIUM_FONTSIZE, label=s, figsize=(17, 3))
    ax.set_title('Average trip duration (min)', fontsize=LARGE_FONTSIZE)
    ax.set_xticklabels(day_of_week_mapping.values());
    ax.set_ylabel('Trip duration\n(Minutes)', fontsize=MEDIUM_FONTSIZE)
plt.legend()
#Display the number of trips by day of the week
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(17, 6))
colors = ['b', 'g']
for i, s in enumerate(['Subscriber', 'Customer']):
    total_trips_by_day = sorted_data[sorted_data['Subscription Type']==s].groupby('Day_of_the_week')['Trip ID'].count()
    total_trips_by_day.plot(ax=axes[i], label=s, color= colors[i], 
                            legend=True, fontsize=MEDIUM_FONTSIZE)
    axes[i].set_title('Total trips by day of the week\n({})'.format(s), fontsize=LARGE_FONTSIZE)
    axes[i].set_xticklabels(day_of_week_mapping.values());
    axes[i].set_ylabel('Number of trips', fontsize=MEDIUM_FONTSIZE);
plt.tight_layout()
plt.figure()
sorted_data['Subscription Type'].value_counts().plot(kind='barh', fontsize=MEDIUM_FONTSIZE)
plt.xlabel('# of riders', fontsize=MEDIUM_FONTSIZE)
plt.title('Rider types', fontsize=LARGE_FONTSIZE)
````

Subscribers make up 79% of Bay Area Bike users, and their trip durations are constant throughout the week, with a very slight increase over the weekend. On the other hand, customers tend to rent bikes on the weekend and ride for 4 times longer than subscribers.

## A Chord Diagram of Bay Area Bike trips

Let's visualize the trips using a Chord diagram as it provides a quick means of viewing traffic flow within the system. This plot is based on Shaun Jacobsen's excellent [Divvy Spokes](http://transitized.com/divvyspokes/) graphic. Scrolling over each station in the plot shows the destinations and number of trips originating from that station.

Each station is color-coded according to its city as reported by Bay Area Bikes.

Color code:
* Blue: San Francisco
* Green: San Jose
* Red: Redwood City
* Orange: Mountain View
* Purple: Palo Alto

````python
from collections import defaultdict
station_names = sorted_data['Start Station'].unique()
station_names.sort()
#Store all the trips in a data structure
#Filter by Start Terminal
trips_counter = []
THRESHOLD = 0
#Sort the stations alphabetically
for i, (station_id, station) in enumerate(sorted_data.sort_index(by='Start Station')\
                                          .groupby('Start Station')):
    #use a counter object to track the number of trips to each destination
    destination_counts = defaultdict(lambda: 0)
    for _, trip in station.sort_index(by='Start Station').iterrows():
        #Count the number of trips with this destination
        destination_counts[trip['End Station']] += 1
    trips_counter.append([destination_counts[s] for s in station_names])
#Store all the data in json format
with open('trip_counts.json', 'a') as f:
    json.dumps(trips_counter, f)
#Load the station data
station_data = pd.read_csv('201402_station_data.csv')
#Sort the stations in alphabetical order 
station_data = station_data.sort_index(by='name')
#Define a colormap for each station
color_map = ['#377eb8', '#4daf4a','#e41a1c', '#984ea3', '#ff7f00']
city_to_colormap = dict(zip(station_data.landmark.unique(), color_map))
#Add a color column to the dataframe
station_data['color'] = [city_to_colormap[s] for s in station_data.landmark]
#Write the data to a file for passing to our D3 plot
station_data[['name', 'color']].to_csv('station_names.csv', index=False)
````
